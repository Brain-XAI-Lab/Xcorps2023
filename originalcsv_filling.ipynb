```python
# Preprocessing Part 1 - Data cutting (Label 사잇값 제거)
def filling(df):

    drops = []
    label = df.iloc[31,:]


    for i in range(1,12):
        globals()["index_{}".format(i)] = label[label == i].index.tolist()

    merged_index = index_1+index_2+index_3+index_4+index_5+index_6+index_7+index_8+index_9+index_10+index_11
    merged_index.sort()
    merged_index = list(map(int, merged_index))

    label[index_1[0]:index_1[1]] = 1  # 1로 마킹된 라벨을 채움.

    # 2 부터 11까지 마킹된 라벨을 채움.
    for i in range(2, 12):
        for t in range(0, 20, 2):
            index_start = int(globals()["index_{}".format(i)][t])  # 인덱스를 정수형으로 변환
            index_end = int(globals()["index_{}".format(i)][t + 1])  # 인덱스를 정수형으로 변환
            label[index_start:index_end + 1] = i


    df = df.drop(range(merged_index[-1],len(df.T)),axis=1)       # 마지막 필요 없는 0을 삭제.

    for i in range(0, len(merged_index)-2, 2):  # for 문을 이용하여 merged_index를 두 단계씩 이동.
        del_start = merged_index[i+1] + 1  # 삭제를 시작할 인덱스를 설정.
        del_end = merged_index[i+2]  # 삭제를 끝낼 인덱스를 설정.
        drops.extend(range(del_start, del_end))  # 삭제할 인덱스를 리스트에 추가.

    df = df.drop(drops,axis=1)  # 삭제할 행을 한 번에 모두 삭제

    df = df.T.reset_index()
    df = df.drop('index', axis=1)
    df = df.T
    return df
```

```python
data = filling(dfinit)
```

```python
# Train EEG datas with ESN Model
# train_data : all columns between row = 1 to 17
# target_data : all columns when row = 31

train_data = data.iloc[1:17, :]
target_data = data.iloc[31, :]

x_train, x_test, y_train, y_test = train_test_split(train_data, target_data, test_size=0.2, shuffle=False, random_state=42)

esn = ESN()
esn.fit(x_train, 12, y_train)
pred = esn.predict(x_test)

print(pred)
```
